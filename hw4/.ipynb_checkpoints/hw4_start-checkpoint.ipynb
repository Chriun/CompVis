```python
'''
The following code for downloading, importing and displaying the Fashion MNIST dataset is adapted
from ChatGPT, accessed February 27, 2024
'''
import numpy as np
import requests
import gzip
import os

def download_fashion_mnist(url, file_name):
    # Check if file already exists
    if not os.path.exists(file_name):
        # Download the file
        r = requests.get(url)
        with open(file_name, 'wb') as f:
            f.write(r.content)

def load_fashion_mnist(image_file, label_file):
    # Read image file
    with gzip.open(image_file, 'rb') as f:
        images = np.frombuffer(f.read(), np.uint8, offset=16).reshape(-1, 28, 28)

    # Read label file
    with gzip.open(label_file, 'rb') as f:
        labels = np.frombuffer(f.read(), np.uint8, offset=8)

    return images, labels

# URLs for Fashion MNIST dataset
image_url = 'http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/train-images-idx3-ubyte.gz'
label_url = 'http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/train-labels-idx1-ubyte.gz'

# File names to save the downloaded files
image_file = 'train-images-idx3-ubyte.gz'
label_file = 'train-labels-idx1-ubyte.gz'

# Download Fashion MNIST dataset
download_fashion_mnist(image_url, image_file)
download_fashion_mnist(label_url, label_file)

# Load Fashion MNIST dataset
train_images, train_labels = load_fashion_mnist(image_file, label_file)

# URLs for Fashion MNIST test dataset
test_image_url = 'http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/t10k-images-idx3-ubyte.gz'
test_label_url = 'http://fashion-mnist.s3-website.eu-central-1.amazonaws.com/t10k-labels-idx1-ubyte.gz'

# File names to save the downloaded files for test dataset
test_image_file = 't10k-images-idx3-ubyte.gz'
test_label_file = 't10k-labels-idx1-ubyte.gz'

# Download Fashion MNIST dataset
download_fashion_mnist(test_image_url, test_image_file)
download_fashion_mnist(test_label_url, test_label_file)

# Load Fashion MNIST dataset
test_images, test_labels = load_fashion_mnist(test_image_file, test_label_file)

# Class labels
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
'''Outputting Examples'''
from matplotlib import pyplot as plt

#Random selection of imgs in notebooks + labels/class names
for i in range(5):
    num, im_h, im_w = train_images.shape
    rand = np.random.randint(num)
    im1 = train_images[rand]
    im1Label = train_labels[rand]
    plt.imshow(im1[:, :])
    plt.gray()
    plt.yticks([])
    plt.xticks([])
    plt.xlabel(class_names[im1Label])
    plt.show()
```

```python
'''Statistics'''
#counts all occurrences of each label
ctr = np.bincount(train_labels)
ctr_str = np.array(ctr, dtype=str)
print("Training Set")
for i in range(ctr.shape[0]):
    print(str(class_names[i]) + ": " + str(ctr[i]))
print()
ctr = np.bincount(test_labels)
print("Test Set")
for i in range(ctr.shape[0]):
    print(str(class_names[i]) + ": " + str(ctr[i]))
```

```python
'''Normalization + Sample Test'''
#normalization
def normalize(imgs):
    imgs = imgs.reshape(imgs.shape[0], -1)
    means = np.mean(imgs)
    norm = imgs-means
    std = np.std(norm)
    return norm/std

test_norm = normalize(test_images)
train_norm = normalize(train_images)

#sample KNeighbors
from sklearn.neighbors import KNeighborsClassifier
sample_size = 25
samples = np.random.choice(test_norm.shape[0], size=sample_size, replace=False)
test_samples = test_norm[samples]
test_samples_labels = test_labels[samples]

test = KNeighborsClassifier(n_neighbors=5)
test.fit(test_norm, test_labels)
pred = test.predict(test_samples)

#printing
for i in range(sample_size):
    print("Test Image", samples[i])
    if (pred[i] == test_samples_labels[i]):
        print("\tCorrectly Predicted Label:", class_names[test_samples_labels[i]])
        print()
    else:
        print("\tPredicted:", class_names[pred[i]], end=", ")
        print("Correct:", class_names[test_samples_labels[i]])
        print()
```

```python
#Calculates accuracy by summing up the instances when the two values are the same
def compute_accuracy(test_pairs, num_classes):
    test_pairs_np = np.array(test_pairs)
    correct = np.sum(test_pairs_np[:, 0] == test_pairs_np[:, 1])
    return correct / len(test_pairs)

#Calculates per class accuracy via bincount
#Finds all occurrences of the true labels, then finds the instances when the two values are the same,
#and finds those occurrences again
def compute_per_class_accuracy(test_pairs, num_classes):
    test_pairs_np = np.array(test_pairs)
    true_labels = test_pairs_np[:, 0]
    predicted = test_pairs_np[:, 1]
    classes = np.bincount(true_labels)
    correct = np.bincount(true_labels[true_labels == predicted])
    
    return correct/classes

#Calculates confusion matrix via bincount
#Calculates the occurrence in unraveled form before reshaping into n x n matrix
def compute_confusion_matrix(test_pairs, num_classes):
    test_pairs_np = np.array(test_pairs)
    confusion = np.bincount(test_pairs_np[:, 0] * num_classes + test_pairs_np[:, 1]).reshape(num_classes, num_classes)
    return confusion
```

```python
'''
Test code for error metrics. Leave this unchanged.
'''
test_pairs = [
    (1, 1),
    (1, 4),
    (0, 0),
    (0, 2),
    (3, 1),
    (4, 2),
    (4, 4),
    (2, 2),
    (2, 2),
    (3, 3),
    (4, 0),
    (4, 4),
    (1, 1),
    (0, 0),
    (0, 2),
    (1, 4),
    (1, 1),
    (0, 0),
    (0, 1),
    (2, 3),
]

num_classes = 5
accuracy = compute_accuracy(test_pairs, num_classes)
print(f'accuracy: {accuracy:.2f}')

per_class_accuracy = compute_per_class_accuracy(test_pairs, num_classes)
print()
print('Per class accuracy')
for i, acc in enumerate(per_class_accuracy):
    print(f'{i}: {acc:4.2f}')

cm = compute_confusion_matrix(test_pairs, num_classes)
print(f'\nConfusion matrix')
for i in range(num_classes):
    print(f'{i:2d}:', end='')
    for j in range(num_classes):
        print(f' {cm[i, j]:2d}', end='')
    print()
```

```python
'''Tuning'''
#Setting 1000 train imgs as validation set, the rest as training set
valid = np.random.choice(test_norm.shape[0], size=1000, replace=False)
validImgs = train_norm[valid]
validLabels = train_labels[valid]
mask = np.ones(train_norm.shape[0], dtype=bool)
mask[valid] = 0
trainImgs = train_norm[mask]
trainLabels = train_labels[mask]

#Parameters
n_neighbors = np.arange(1, 21)

weights = ['uniform', 'distance']

best_accuracy = 0
best_params = [0, '']

#Loops through combinations and finds best accuracy
for n in n_neighbors:
    for weight in weights:
        knn = KNeighborsClassifier(n_neighbors = n, weights=weight)
        knn.fit(trainImgs, trainLabels)
        pred = knn.predict(validImgs)
        pairs = np.column_stack((pred, validLabels)).astype(int)
        accuracy = compute_accuracy(pairs, len(class_names))
        print("N_Neighbors: " + str(n) + ", Weight: " + weight)
        print("\tOverall Accuracy:", accuracy)
        print()
        if (accuracy > best_accuracy):
            best_accuracy = accuracy
            best_params = [n, weight]
print("Best Parameters: N_Neighbors = " + str(best_params[0]) + ", Weight = " + best_params[1])
```

```python
'''Final Test'''
knn = KNeighborsClassifier(n_neighbors=best_params[0], weights=best_params[1])
knn.fit(train_norm, train_labels)
pred = knn.predict(test_norm)
pairs = np.column_stack((pred, test_labels)).astype(int)

num_classes = len(class_names)
print("Corresponding Labels")
for i in range(num_classes):
    print(str(i) + ": " + class_names[i])
print()
    
accuracy = compute_accuracy(pairs, len(class_names))
print(f'Overall Accuracy: {accuracy:.2f}')

per_class_accuracy = compute_per_class_accuracy(pairs, num_classes)
print()
print('Per class accuracy')
for i, acc in enumerate(per_class_accuracy):
    print(f'{i}: {acc:4.2f}')
    
cm = compute_confusion_matrix(pairs, num_classes)
print(f'\nConfusion matrix')
for i in range(num_classes):
    print(f'{i:2d}:', end='')
    for j in range(num_classes):
        print(f' {cm[i, j]:2d}', end='')
    print()
```

Discussion:
The classes that are the easiest to recognize are the trousers, sandals, and bags with per class accuuracy being the closest to 1. The hardest to recognize are the tops, pullovers, and shirts. This makes sense considering those three are the classes that are the most similar and therefore easily confused with each other. I was very surprised by how well it classified sandals vs sneakers and ankle boots since they are also a bit similar. The most important parameter is definitely the nearest neighbors parameter because as I increased the neighbors to above 20s, the accuracy began to drastically fall.
